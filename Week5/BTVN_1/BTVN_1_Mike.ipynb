```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#Task 1: Sắp xếp và in ra 10 thành phố có dân số lớn nhất & 10 thành phố có dân số nhỏ nhất

```python
df['Country'] = df['Country'].str.replace('*', '')
df['Country'] = df['Country'].str.strip()
```

```python
df['Population'] = df['Population'].str.replace(',', '').astype(float)
df.sort_values(by='Population', ascending=False, inplace=True)
```

```python
#10 thành phố có dân số lớn nhất
print(df.head(10))
```

```python
#10 thành phố có dân số nhỏ nhất
print(df.tail(10))
```

Task 2: In ra tên các quốc gia có tối thiểu 3 thành phố trong danh sách này

```python
count_cities = df.groupby('Country')['City'].count().reset_index()
count_cities = count_cities[count_cities['City'] >= 3]
count_cities.sort_values(by='City', ascending=False, inplace=True)
print(count_cities)
```

Task 3: In ra Top 5 quốc gia có nhiều thành phố xuất hiện trong bảng này nhất

```python
print(count_cities.head(5))
```

Task 4: In ra các thành phố có dân số & diện tích đều nằm trong Top 20

```python
df['Area KM2'] = pd.to_numeric(df['Area KM2'], errors='coerce')
df = df[df['Area KM2'].isna() == False]
df.info()
```

```python
df.sort_values(by=['Population','Area KM2'], ascending=False, inplace=True)
df.head(20)
```

Task 5: Thống kê mật độ dân số theo quốc gia

```python
population_countries = df.groupby('Country')['Population'].sum().reset_index()
population_countries.sort_values(by='Population', ascending=False, inplace=True)
population_countries
```

Task 6: Tìm các thành phố có dân số lớn nhất của từng quốc gia (chỉ tính của những quốc gia có 2 thành phố xuất hiện trở lên trong bảng)

```python
count_cities = df.groupby('Country')['City'].count().reset_index()
count_cities = count_cities[count_cities['City'] >= 2]
qualified_countries = count_cities['Country'].unique()
qualified_countries
```

```python
df = df[df['Country'].isin(qualified_countries)]
df
```
