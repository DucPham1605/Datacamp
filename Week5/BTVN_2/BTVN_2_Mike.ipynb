```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('dataset_2.csv')
```

```python
df.info()
df.shape
```

Task 1: In ra index của 10 học sinh có math score và reading score đều nằm trong Top 100

```python
top_math_scores = df['math score'].nlargest(100)
top_reading_scores = df['reading score'].nlargest(100)

top_students = df[(df['math score'].isin(top_math_scores)) & (df['reading score'].isin(top_reading_scores))]
top_students.shape
```

```python
top_students.index[:10]
```

Task 2: In ra index của 20 học sinh có điểm writing score cao nhất của mỗi group (race/ethnicity)

```python
top_writing_scores = df.groupby('race/ethnicity').apply(
    lambda x: x.nlargest(20,'writing score')
)

print(top_writing_scores.index)
```

Task 3: Thống kê số lượng học sinh của group A theo parental level of education

```python
students_group_A = df[df['race/ethnicity'] == 'group A'].groupby('parental level of education')['parental level of education'].agg(
    num_students = 'count'
).reset_index()
students_group_A.sort_values(by='num_students', ascending=False)
```

Task 4: Có phải xu hướng chung, cha mẹ có học vấn càng cao thì điểm số trung bình 3 môn của con cái cũng càng cao ko

```python
df['average score'] = df.apply(lambda x: np.mean([x['math score'], x['reading score'], x['writing score']]), axis=1)
```

```python
student_avg_score = df.groupby('parental level of education')['average score'].mean().sort_values(ascending=False)
student_avg_score
```

Khi parental level of education là dạng Ordinal data với thứ tự tăng dần từ some high school đến master's degree, dựa vào bảng trên, ta có thể nhận thấy khi cha mẹ càng có học vấn cao thì điểm số trung bình 3 môn của con cũng càng cao

Task 5: Chất lượng bữa ăn (lunch) có tương phản với điểm số trung bình 3 môn của học sinh không

```python
avg_score_lunch = df.groupby('lunch')['average score'].mean().sort_values(ascending=False)
avg_score_lunch
```

Dựa vào kết quả trên, ta có thể thấy học sinh có chất lượng bữa ăn thuộc tiêu chuẩn thường có điểm số trung bình cao hơn, trong khi học sinh có free/reduced meal thì có điểm trung bình kém hơn.


Task 6: Tìm ra Top 10 học sinh có điểm toán cao nhất của mỗi group

```python
top_10_math_scores = df.groupby('race/ethnicity').apply(
    lambda x: x.nlargest(10,'math score')
)

top_10_math_scores
```

Task 7: Với những học sinh ở cùng 1 group, cùng parental level of education, cùng loại bữa ăn (lunch), việc tham gia test preparation course có giúp học sinh đó có điểm trung bình 3 môn cao hơn những học sinh không tham gia ko

```python
task_7 = df.groupby(['race/ethnicity','parental level of education','lunch','test preparation course'])['average score'].mean().reset_index()
task_7.sort_values(by='race/ethnicity', ascending=False)
task_7
```

Vì có khá nhiều observations nên em không visualize kết quả để đánh giá tổng thể được. Tuy nhiên khi kiểm tra kết quả trên thì ta thấy nhìn chung việc tham test preparation course có giúp học sinh đó đạt điểm trung bình cao hơn, chỉ riêng có một số trường hợp nhỏ lẻ thì không như vậy.
