```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('dataset_3.csv')
df.info()
df.shape
```

```python
df['revenue'] = df.apply(lambda row: row['retail_price']*row['units_sold'], axis=1)
top_products = df[df['revenue'] > 100000].sort_values(by='revenue', ascending=False)
top_products
```

Task 2: Liệt kê các sản phẩm có units_sold nằm trong top 100, đồng thời nằm trong nhóm 10% sản phẩm có rating cao nhất

```python
top_units_sold = df['units_sold'].nlargest(100)
top_rated_products =df['rating'].nlargest(157)

top_products = df[(df['units_sold'].isin(top_units_sold) & (df['rating'].isin(top_rated_products)))]
top_products
```

Task 3: Có phải sản phẩm càng được rating nhiều thì càng bán được nhiều sản phẩm ko

```python
correlation = df['rating_count'].corr(df['units_sold'])
correlation
```

Hệ số tương quan giữa số lần rate và số sản phẩm được bán nhiều cao, từ đó cho thấy sự tương quan mạnh giữa hai yếu tố này. Vì vậy có thể nhận định rằng sản phẩm càng được rated nhiều thì càng bán được nhiều sản phẩm

Task 4: Thống kê Top 50 sản phẩm nằm trong rating từ 3.0 → 4.0 bán chạy nhất

```python
top_50_products = df[(df['rating'] >= 3.0) & (df['rating'] <= 4.0)]
top_50_products.sort_values(by=['units_sold'],ascending = False,inplace=True)
top_50_products[:50]
```
